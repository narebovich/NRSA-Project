```python
#Census_acs5_api pulls tables from the American Community Survey 5 year basedon year input
#B01003_001E is total population
#B17001_001E is total poverty status
#NAME is State, County, Tract name of geography
#B25081_001E Housing with/without mortgage
#B23001_001E Employment Status
#B25002_003E Vacancy Rate
#B25081_002E Housing with Mortgage
#B25081_008E Housing without mortgage
#B25008_002E Owner Occupied
#B25008_003E Renter Occupied

#Base url for api
host = 'https://api.census.gov/data'
#Dataset we are drawing to data from acs/acs5 is American Community Survey 5-year acs/acs1 is ACS 1-year
dataset = 'acs/acs5'
#Initial year is the first year/only year of data that you want
initial_year ='2019'
#Base_url joins the above varaibles together with / like a url
base_url = '/'.join([host,initial_year,dataset])

#get_vars specifies the numeric table ids of the data you want
#NAME is State, County, Tract name of geography
#B01003_001E is total population
#B17001_002E is number of people below poverty line
#B19013_001E is the median income of a tract
#B25002_003E is number of total vacant properties
#B25081_002E housing with mortgage
#B25081_008E housing without mortgage
#B25008_002E 
#S0101_C01_033E median age
get_vars = ['NAME','B17001_002E','B01003_001E','B19013_001E', "B25002_003E", 'B25081_002E','B25081_008E',
            'B25008_002E','B25008_003E']

#Predicates indicate the geography you want to search by
predicates = {}
predicates['get'] = ','.join(get_vars)
#For predicate indicates which geography you want
predicates['for'] = 'tract:*'
#In predicates narrow down the search
predicates['in'] = ['state:20','county:209']


#Response requests the data from the url outlined above using the predicate dictionary as its parameters
response = requests.get(base_url,params = predicates)

#Below prints the url address created and the status code of the response. A successful response is 200.
print("API URL")
print(response.url)
print('\n')
print("Status Code")
print(response.status_code)

#Creating a pandas dataframe based in the initial year of data
col_names = ['Name','Poverty Population '+ initial_year,
             'Population ' + initial_year,'Median Income ' + initial_year,'Number of Vacants ' + initial_year,
             'Housing with Mortgage ' + initial_year, 'Housing without Mortgage ' + initial_year, 
             'Owner Occupied ' + initial_year, 'Renter Occupied ' + initial_year,
             'State','County','Tract']
tract_nrsa_data_initial = pd.DataFrame(data = response.json()[1:], columns = col_names)
#Reorder the columns to make it more readable
tract_nrsa_data = tract_nrsa_data_initial [["Name",'State','County','Tract','Poverty Population '+ initial_year,
             'Population ' + initial_year,'Median Income ' + initial_year,'Number of Vacants ' + initial_year,
             'Housing with Mortgage ' + initial_year, 'Housing without Mortgage ' + initial_year, 
             'Owner Occupied ' + initial_year, 'Renter Occupied ' + initial_year]]
```
